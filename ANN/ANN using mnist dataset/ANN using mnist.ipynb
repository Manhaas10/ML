```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0

# Flatten the 28x28 images into 784-dimensional vectors
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# Convert labels to categorical format
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
km=Sequential([
    Flatten(input_shape=(784,)),
    Dense(64,activation='relu'),
    Dense(64,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
# help(km.compile)
km.compile(optimizer='adam',
           loss='categorical_crossentropy',
           metrics=['accuracy']
          )
km.fit(x_train,y_train,epochs=10)
```

```python
rain_loss, train_accuracy = km.evaluate(x_train, y_train, verbose=0)
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")

# Evaluate the model on test data
test_loss, test_accuracy = km.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
def predict(model, x):
    predictions = model.predict(x)
    return np.argmax(predictions, axis=1)  # Returns the class with the highest probability

# Test the predict function on a sample
sample_predictions = predict(km, x_test[:10])
print("Sample Predictions:", sample_predictions)
print("Actual Labels:", np.argmax(y_test[:10], axis=1))
```
